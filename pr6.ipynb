```python
print("КАФЕ")

menu = []

orders = []

def add_dish():
    """Добавить блюдо в меню"""
    name = input("Название блюда: ")
    price = int(input("Цена: "))
    category = input("Категория (суп/салат/горячее/напиток): ")

    menu.append({
        "name": name,
        "price": price,
        "category": category
    })
    print(f"✅ {name} добавлено!")

def make_order():
    """Сделать заказ"""
    table = input("Номер стола: ")

    print("Меню:")
    for i, dish in enumerate(menu):
        print(f"{i+1}. {dish['name']} - {dish['price']} руб.")

    dish_numbers = input("Введите номера блюд через пробел: ").split()

    order_dishes = []
    total = 0

    for num in dish_numbers:
        idx = int(num) - 1
        dish = menu[idx]
        order_dishes.append(dish['name'])
        total += dish['price']

    orders.append({
        "table": table,
        "dishes": order_dishes,
        "total": total,
        "status": "готовится"
    })

    print(f"✅ Заказ принят! Сумма: {total} руб.")

def show_popular():
    """Показать популярные блюда"""
    dish_count = {}

    for order in orders:
        for dish in order['dishes']:
            if dish in dish_count:
                dish_count[dish] += 1
            else:
                dish_count[dish] = 1

    print("Популярные блюда:")
    for dish, count in dish_count.items():
        print(f"{dish}: заказали {count} раз")

def show_stats():
    """Показать статистику по категориям"""
    cat_total = {}

    for dish in menu:
        cat = dish['category']
        if cat in cat_total:
            cat_total[cat] += dish['price']
        else:
            cat_total[cat] = dish['price']

    print("Выручка по категориям:")
    for cat, total in cat_total.items():
        print(f"{cat}: {total} руб.")

while True:
    print("\n1. Добавить блюдо")
    print("2. Сделать заказ")
    print("3. Популярные блюда")
    print("4. Статистика")
    print("5. Выход")

    choice = input("Выбери действие: ")

    if choice == "1":
        add_dish()
    elif choice == "2":
        make_order()
    elif choice == "3":
        show_popular()
    elif choice == "4":
        show_stats()
    elif choice == "5":
        print("Пока!")
        break
```

```python
print("Успеваемость студентов")

students = []

grades = []

def add_student():
    """Добавить студента"""
    print("\n--- Новый студент ---")
    id = input("ID студента: ")
    name = input("ФИО: ")
    group = input("Группа: ")

    students.append({
        "id": id,
        "name": name,
        "group": group
    })
    print(f"✅ Студент {name} добавлен!")

def add_grade():
    """Добавить оценку"""
    print("\n--- Новая оценка ---")

    print("Студенты:")
    for s in students:
        print(f"{s['id']}. {s['name']}")

    student_id = input("ID студента: ")
    subject = input("Предмет: ")
    grade = int(input("Оценка (2-5): "))

    grades.append({
        "student_id": student_id,
        "subject": subject,
        "grade": grade
    })
    print(f"✅ Оценка {grade} добавлена!")

def student_avg():
    """Средний балл студента"""
    student_id = input("ID студента: ")

    student_name = ""
    for s in students:
        if s['id'] == student_id:
            student_name = s['name']
            break

    total = 0
    count = 0
    for g in grades:
        if g['student_id'] == student_id:
            total += g['grade']
            count += 1

    if count > 0:
        avg = total / count
        print(f"{student_name}: средний балл = {avg:.2f}")
    else:
        print("У студента нет оценок")

def find_bad_students():
    """Найти студентов с плохими оценками"""
    min_grade = int(input("Ниже какого балла ищем? (например 3): "))

    bad_students = set()

    for g in grades:
        if g['grade'] < min_grade:
            bad_students.add(g['student_id'])

    print("Студенты с низкими оценками:")
    for student_id in bad_students:
        for s in students:
            if s['id'] == student_id:
                print(f"{s['name']} ({s['group']})")

def show_rating():
    """Рейтинг студентов"""

    rating = []

    for student in students:
        total = 0
        count = 0
        for g in grades:
            if g['student_id'] == student['id']:
                total += g['grade']
                count += 1

        if count > 0:
            avg = total / count
            rating.append({
                "name": student['name'],
                "group": student['group'],
                "avg": avg
            })

    rating.sort(key=lambda x: x['avg'], reverse=True)

    print("\n=== РЕЙТИНГ СТУДЕНТОВ ===")
    for i, s in enumerate(rating):
        print(f"{i+1}. {s['name']} ({s['group']}) - {s['avg']:.2f}")

while True:
    print("\n1. Добавить студента")
    print("2. Добавить оценку")
    print("3. Средний балл студента")
    print("4. Найти должников")
    print("5. Показать рейтинг")
    print("6. Выход")

    choice = input("Выбери действие: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        add_grade()
    elif choice == "3":
        student_avg()
    elif choice == "4":
        find_bad_students()
    elif choice == "5":
        show_rating()
    elif choice == "6":
        print("Пока!")
        break
```
